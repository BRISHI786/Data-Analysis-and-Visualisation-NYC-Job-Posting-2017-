### : __Graph Analysis For Selected Data__

### - _Line Graph_

### 1 . _Salary Range Starts From_ :

```python
df['Salary Range From'].plot(figsize=(14,6))
plt.title("----------      ******      Salary Range From       ******     ----------")
plt.show()
```

### :  __Job Count Based On Job Position__ 

```python
df['Posting Type'].value_counts().plot(kind='bar')
plt.title("-------      ****   Total Job Count Based On Job Position     ****     -------")
plt.xlabel('Job Position')
plt.ylabel('Job Count')
plt.show()
```

### :  __NYC Job Agencies__ 

```python
Job_Agency = df['Agency'].value_counts()
Top_Agency = Job_Agency[:10]
Top_Agency.plot(kind='barh',alpha=0.6,figsize=(12,10))
plt.title("----------    ******   NYC JOB AGENCIES   ******   ----------")
plt.show()
```

### : __Job - Hours / Shift__

```python
Business = df['Hours/Shift'].value_counts()
new_var = Business[:5]
new_var.plot(kind='bar',alpha=0.4,figsize=(10,5))
plt.title("-------      ****   JOB : HOURS / SHIFT    ****     -------")
plt.show()
```

### : __Star Style Graph Plotting__
